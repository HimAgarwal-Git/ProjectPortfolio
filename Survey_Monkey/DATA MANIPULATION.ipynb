# Combined the entries using demographic vs variable 


```python
dataset_melted=dataset_modified.melt(id_vars=id_vars,value_vars=value_vars,var_name="questions",value_name="answers_recorded")
```

### DID ANALYSIS TO FIGURE OUT HOW MANY QUESTIONS WERE ASKED AND WHO ALL ANSWERED

```python
questions_only=pd.read_excel(file+"\\Downloads\Data - Survey Monkey Output.xlsx",sheet_name="Sheet2")
```

```python
dropset=['RAW', 'Valid Answers','Unnamed: 3',
        'Respondent ID -- ']
```

### PERFORMED SEQUEL JOIN USING PANDAS

```python
dataset_merged=pd.merge(left=dataset_melted, right=questions_only, how="left", left_on="questions", right_on="Questions+ Valid Answers")
```

```python
print("original dataset",len(dataset_melted))
print("merged dataset",len(dataset_merged))
```

```python
dataset_merged
```

```python
dataset_merged_final=dataset_merged.copy()
```

```python
dataset_merged.groupby("Questions")['Respondent ID -- '].nunique().reset_index()
```

```python
respondents=dataset_merged['answers_recorded'].dropna()
```

```python
respondents
```

```python
respondents=dataset_merged[dataset_merged["answers_recorded"].notna()]
respondents
```

```python
respondents=respondents.groupby("Questions")["Respondent ID -- "].nunique().reset_index()
```

```python
respondents.rename(columns={"Respondent ID --":"Respondents"},inplace=True)
```

```python
respondents
```

```python
dataset_mergedTwo=pd.merge(left=dataset_merged,right=respondents,how="left",left_on="Questions",right_on="Questions")
print(len(dataset_merged),len(dataset_mergedTwo))
```

## ANALYSIS 2: NUMBER OF SAME ANSWERES OR UNIQUE ANSWERS IN THE POPULATION
